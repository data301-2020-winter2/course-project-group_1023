```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import datetime
import sys
sys.path.append("..")
```

```python
#### Importing function from the "Scripts" file in the analysis folder 

from Scripts.project_functions import load_and_process
data=load_and_process('../../data/raw/weatherAUS.csv')
```

Creating a dataframe that only consists of perth

```python
### getting data only in perth
data_in_perth=  data[data["Location"]=='Perth']
```

```python
### adding a new column so that the dataes are in dateime format 
data_in_perth['Datetime'] = pd.to_datetime(data_in_perth['Date']).dt.year
```

```python
data_in_perth['Datetime_month'] = pd.to_datetime(data_in_perth['Date'])
```

```python
data_in_perth['Datetime_day'] = pd.to_datetime(data_in_perth['Date']).dt.day
```

```python
data_in_perth['Month_and_day'] = pd.to_datetime(data_in_perth['Date']).dt.strftime('%m-%d')
```

```python
data_in_perth
```

The dataset above consists of obervations only in the city of Perth. The columns in this specific dataset are the observed values of data in the city of Perth. 

```python
plot_perth = sns.relplot(x="Month_and_day", 
                          y="Temp3pm", 
                          kind="line", 
                          hue="Datetime",
                          aspect=5,
                          data=data_in_perth)

plot_perth.fig.suptitle("Tracing Temperature at 3pm over all Of the dates in the data set (Celsius)")
```

This is a traceplot of the temperatures at 3pm at Perth consisting over the years between 2011-2017. The horizontal axis are months and days and the years are categorized. I inteded to see a difference in temperature change over the years. All years follow a similar path of temperature at 3pm of a day.

```python
### box plot with temperature at 3pm

Perth_box = sns.boxplot(x="Datetime", y="Temp3pm", data=data_in_perth).set_title("Temperature at 3pm based on each year (Celsius)")
```

This is a boxplot of the temperatures at 3pm at Perth over the years. As we can see from 2010 to 2017 the average temperature at 3pm has not changed significantly over time. 

```python
### box plot with max temperatures 
Perth_max = sns.boxplot(x="Datetime", y="MaxTemp", data=data_in_perth).set_title("Maximum temperature boxplot over the years (Celsius)")
```

This is a boxplot of the maximum temperatures over the years in Perth. Again, the maximum temperatures at Perth does not show a difference in the averages over time. 

```python
### box plot with min temeperatures 
Perth_max = sns.boxplot(x="Datetime", y="MinTemp", data=data_in_perth).set_title("Minimum Temperature boxplot over the years (Celsius)")
```

This is a boxplot of the minimum temperatures over the years in Perth. The minimum temperatures at Perth also did not change significanlty over time. Although an observation can be noticed that 2008 has a significantly different teperature change compared to the other years, not just for Max Temperature but in general. This can indicate that the sample size of the data set is insufficent to observe climate change. Even though this year can be initially looks like an outlier, the difference in the average temperature isn't too big for it to be considered an outlier. In addition these box plots are created with a minimum of 365 data points, making it not able for it to be disqualified.
