```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import datetime
```

```python
from scripts import project_functions # This is called a relative import
data = project_functions.load_and_process('../../data/raw/weatherAUS.csv')
```

```python
### getting data only in sydney 
data_in_sydney=  data[data["Location"]=='Sydney']
```

```python
### adding a new column so that the dataes are in dateime format 
data_in_sydney['Datetime'] = pd.to_datetime(data_in_sydney['Date']).dt.year
```

```python
data_in_sydney['Datetime_month'] = pd.to_datetime(data_in_sydney['Date'])
```

```python
data_in_sydney['Datetime_day'] = pd.to_datetime(data_in_sydney['Date']).dt.day
```

```python
data_in_sydney['Month_and_day'] = pd.to_datetime(data_in_sydney['Date']).dt.strftime('%m-%d')
```

```python
data_in_sydney
```

```python
plot_sydney = sns.relplot(x="Month_and_day", 
                          y="Temp3pm", 
                          kind="line", 
                          hue="Datetime",
                          aspect=5,
                          data=data_in_sydney)

plot_sydney.fig.suptitle("Tracing Temperature at 3pm over all Of the dates in the data set")
```

```python
### box plot with temperature at 3pm

Sydney_box = sns.boxplot(x="Datetime", y="Temp3pm", data=data_in_sydney).set_title("Temperature at 3pm based on each year")
```

```python
### box plot with max temperatures 
Sydney_max = sns.boxplot(x="Datetime", y="MaxTemp", data=data_in_sydney).set_title("Maximum temperature boxplot over the years")
```

```python
### box plot with min temeperatures 
Sydney_max = sns.boxplot(x="Datetime", y="MinTemp", data=data_in_sydney).set_title("Minimum Temperature boxplot over the years")
```

```python
plot_sydney = sns.relplot(x='Date',
                         y= "Percipitation",
                         data=data_in_sydney,
                         aspect = 2)
plot_sydney.fig.suptitle("Plot of rainfall over all of the dates")
```
