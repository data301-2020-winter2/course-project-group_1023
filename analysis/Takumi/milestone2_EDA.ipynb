```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import datetime
```

#### Importing function from the "Scripts" file in the analysis folder 

```python
from Scripts.project_functions import load_and_process
data=load_and_process('../../data/raw/weatherAUS.csv')
```

Creating a dataframe that only consists of sydney 

```python
### getting data only in sydney 
data_in_sydney=  data[data["Location"]=='Sydney']
```

```python
### adding a new column so that the dataes are in dateime format 
data_in_sydney['Datetime'] = pd.to_datetime(data_in_sydney['Date']).dt.year
```

```python
data_in_sydney['Datetime_month'] = pd.to_datetime(data_in_sydney['Date'])
```

```python
data_in_sydney['Datetime_day'] = pd.to_datetime(data_in_sydney['Date']).dt.day
```

```python
data_in_sydney['Month_and_day'] = pd.to_datetime(data_in_sydney['Date']).dt.strftime('%m-%d')
```

```python
data_in_sydney
```

The dataset above consists of obervations only in the city of Sydney. The columns in this specific dataset are the observed values of data in the city of Sydney. 

```python
plot_sydney = sns.relplot(x="Month_and_day", 
                          y="Temp3pm", 
                          kind="line", 
                          hue="Datetime",
                          aspect=5,
                          data=data_in_sydney)

plot_sydney.fig.suptitle("Tracing Temperature at 3pm over all Of the dates in the data set (Celsius)")
```

This is a traceplot of the temperatures at 3pm at Sydney consisting over the years between 2011-2017. The horizontal axis are months and days and the years are categorized. I inteded to see a difference in temperature change over the years. 

```python
### box plot with temperature at 3pm

Sydney_box = sns.boxplot(x="Datetime", y="Temp3pm", data=data_in_sydney).set_title("Temperature at 3pm based on each year (Celsius)")
```

This is a boxplot of the temperatures at 3pm at sydney over the years. As we can see from 2010 to 2017 the average temperature at 3pm has not changed significantly over time. 

```python
### box plot with max temperatures 
Sydney_max = sns.boxplot(x="Datetime", y="MaxTemp", data=data_in_sydney).set_title("Maximum temperature boxplot over the years (Celsius)")
```

This is a boxplot of the maximum temperatures over the years in Sydney. Again, the maximum temperatures at Sydney does not show a difference in the averages over time. 

```python
### box plot with min temeperatures 
Sydney_max = sns.boxplot(x="Datetime", y="MinTemp", data=data_in_sydney).set_title("Minimum Temperature boxplot over the years (Celsius)")
```

This is a boxplot of the minimum temperatures over the years in Sydney. The minimum temperatures at Sydney also did not change significanlty over time.

```python
plot_sydney = sns.relplot(x='Date',
                         y= "Percipitation",
                         data=data_in_sydney,
                         aspect = 2)
plot_sydney.fig.suptitle("Plot of rainfall over all of the dates (mm)")
```

This is a plot of the percipitation in Sydney over the years. I intended to see if there was more percipitation over the years maybe due to gloal warming.
